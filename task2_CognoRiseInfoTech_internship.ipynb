```python
from scipy.stats import pearsonr
from scipy.stats import chi2_contingency
```

# we want to see if there's a correlation between FAT AND WEIGHT
- H0: No correlation between rating AND sugars
- H1: there's a correlation between rating AND sugars

```python
plt.scatter(x=df['rating'], y=df['sugars'])
plt.xlabel('rating')
plt.ylabel('sugars')
plt.title('sugars vs rating', fontsize=16)
```

```python
r, p = pearsonr(df['rating'], df['sugars'])
print(p)
```

```python
if p < 0.05:
    print("rejecting H0:No correlation between rating and sugars")
else:
    print("accepting H0:No correlation between rating and sugars")
```

```python
my_rho=np.corrcoef(df['rating'], df['sugars'])
```

```python
my_rho
```

```python
from scipy.stats import spearmanr
```

```python
# spearman test
r, p = spearmanr(df['rating'], df['sugars'])
print(p)
```

```python
if p < 0.05:
    print("rejecting H0:No correlation between rating and sugars")
else:
    print("accepting H0:No correlation between rating and sugars")
```
